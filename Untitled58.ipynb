```python
region= pd.get_dummies(data.region, drop_first=True)
data['sex']= data.sex.map({'male':1,'female':0})
data['smoker']= data.smoker.map({'yes':1,'no':0})
data
```

```python
region
```

```python
data = data.drop('region',1)
new_data= pd.concat([data,region],axis = 1)
new_data
```

```python
data
```

```python
y = data['expenses']
X = data.drop('expenses',1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state =100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
x = sm.add_constant(X)
results=sm.OLS(y_train,X_train).fit()
results.summary()
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
yhat = model.predict(X_test)
model.score(X_test,y_test)
```

```python
plt.title('Relationship between the error terms')
plt.scatter(yhat,y_test)
plt.xlabel('Predictions')
plt.ylabel('Actual')
plt.xlim(0,60000)
plt.ylim(0,60000)
plt.show()
```

```python
sns.displot(y_test- yhat)
```
